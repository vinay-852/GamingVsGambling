```python
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.fantome.penguinisle',
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.NEWEST
    count=3, # defaults to 100
    filter_score_with=5 # defaults to None(means all score)
)

result, _ = reviews(
    'com.fantome.penguinisle',
    continuation_token=continuation_token
)
```

```python
from google_play_scraper import app, Sort, reviews, reviews_all, permissions, search

# Get app from search
def get_appId_from_search(query, lang="en", country="us", n_hits=1):
    result = search(query, lang="en", country="us", n_hits=1)
    return result[0]['appId']
# Get app info
def get_app_info(appId, lang="en", country="us"):
    app_info = app(appId, lang="en", country="us")
    return app_info
print(get_app_info(get_appId_from_search("whatsapp")))
```
